```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import random
```

```python
#change the vin number in obd to match the vin in the vehicle 

#get the data from vehicle json
data_vehicle = open('vehiclev3.json' , 'r')
data1 = json.load(data_vehicle)

    
#get the data from OBD2s json
data_obd = open('obd2.json' , 'r')
data2 = json.load(data_obd)


#changing the vin and the parameterset 
for index, val in enumerate(data2):
    newParamSet = {} 
    randomindex= random.randint(0,len(data1)-1)
    data2[index]['vin'] = data1[randomindex]['vehicleInfo']['vin']
    #a list from this vehicle 
    listRand = random.randint(4, 15)
    for i in range (listRand):
        newParamSet[data1[randomindex]['logInfo']['logParamSet'][i]['shortName']] = random.randint(0,100)
    
    data2[index]['parameterSet'] = newParamSet
    print(newParamSet)
        
    
#save the new data to new json file 
with open('newobd2.json', 'w') as f:
    json.dump(data2, f)

    
```
